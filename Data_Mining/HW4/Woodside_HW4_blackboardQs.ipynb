Forward/Backward Fill Imputation

3.1 Description
Forward and backward fill are imputation methods that replace missing values
by propagating the last or next observed value:

• Forward Fill (ffill): Replaces missing values with the previous non-null
value.

• Backward Fill (bfill): Replaces missing values with the next non-null
value.


```python
'''Forward Fill'''
F = m_df.copy()

X = F[F['series'] == 'EN.ATM.CO2E.PC']
year = [col for col in data.columns if col.startswith('YR')]
year_to_keep = year[:-3]  # Keep all but the last three year columns
X = X[~X[year_to_keep].isna().all(axis=1)]  # Removes countries with all NaNs across the rows

X[year] = X.groupby('Country')[year].ffill()

leading_nans = X[year].isna().any(axis=1).sum()
#print(f"Number of rows with leading NaNs: {leading_nans}")
leading_nan_rows = X[X[year].isna().any(axis=1)]

# print(f"Number of rows with leading NaNs: {leading_nan_rows.shape[0]}")
# print("Rows with leading NaNs:")
# print(leading_nan_rows[year])


X[year] = X[year].fillna(0)
X = X.iloc[:, :-3]
year = [col for col in X.columns if col.startswith('YR')]  

# Remove any rows that still have NaNs (including the last three years)
#X = X.dropna(subset=year, how='any')


# print("Final shape of X:", X.shape)
# print(X[year].isna().sum())
print(X[year].head())
```

```python
'''Back fill'''
bb = m_df.copy()
X = bb[bb['series'] == 'EN.ATM.CO2E.PC'].copy()

year = [col for col in bb.columns if col.startswith('YR')]
year_to_keep = year[:-3]  # Keeps all years except the last three

X = X[~X[year_to_keep].isna().all(axis=1)]
copy_bfill = X.groupby('Country')[year_to_keep].bfill()
X.update(copy_bfill)
print(X[year_to_keep].head())

na_mask = X[year_to_keep].isna().any(axis=1)
print('\n', X[['Country'] + year_to_keep][na_mask])
#print(X[year_to_keep].dtypes)


remaining_nans = X[year_to_keep].isna().sum()
print("Remaining NaNs per year:")
print(remaining_nans)


countries_with_nans = X[X[year_to_keep].isna().any(axis=1)]['Country'].unique()
if len(countries_with_nans) > 0:
    print(f"There are NaN values in the following countries: {countries_with_nans}")
else:
    print("No remaining NaN values in the DataFrame.")
```

K-Nearest Neighbors

4.1 Description
K-Nearest Neighbors (KNN) Imputation fills missing values by finding the knearest records (rows) in the dataset that are similar to the record with missing
data. The missing values are then imputed by averaging the values of the
neighbors

```python
from sklearn.impute import KNNImputer

subset = m_df[m_df['series'] == 'EN.ATM.CO2E.PC'].copy()
year = [col for col in data.columns if col.startswith('YR')]

subset = subset.dropna(subset=year, how='all')

subset[year] = subset[year].apply(pd.to_numeric, errors='coerce')
subset = subset.dropna(axis=1, how='all')
#print('Initital Dataframe:\n', subset[year].loc[[2,3,4]]) 

for idx in [2, 3, 4]:
    if idx not in subset.index:
        new_row = pd.DataFrame({col: [float('nan')] for col in subset.columns})
        new_row['series'] = 'EN.ATM.CO2E.PC'
        new_row.index = [idx]
        subset = pd.concat([subset, new_row])

subset = subset.sort_index()
year = [col for col in subset.columns if col.startswith('YR')]
print('Initial DataFrame for indices (2, 3, 4):\n', subset[year].loc[[2, 3, 4]])

knn = KNNImputer(n_neighbors=1)
df_knn = knn.fit_transform(subset[year])

subset[year] = pd.DataFrame(df_knn, columns=year, index=subset.index)
print('\nEnd result with values filled (for indices 2, 3, 4):\n', subset[year].loc[[2, 3, 4]])

```

### Task: Model Comparison and Evaluation

```python
# So Model 1 would be the original model with the df_simulation?
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#to add to df
df_simulation['I_X3_B'] = (df_simulation['X3'] == 'B').astype(int)
df_simulation['I_X3_C'] = (df_simulation['X3'] == 'C').astype(int)


df_simulation['X1_squared'] = df_simulation['X1'] ** 2
df_simulation['X2_squared'] = df_simulation['X2'] ** 2
df_simulation['X2_I_X3_B'] = df_simulation['X2'] * df_simulation['I_X3_B']
df_simulation['X2_I_X3_C'] = df_simulation['X2'] * df_simulation['I_X3_C']
df_simulation['X1_I_X3_B'] = df_simulation['X1'] * df_simulation['I_X3_B']
df_simulation['X1_I_X3_C'] = df_simulation['X1'] * df_simulation['I_X3_C']

train_df, test_df = train_test_split(df_simulation, test_size=1/3, random_state=42)

# Model 1 (Correct set of predictors)
X_train_model1 = train_df[['X1', 'X2', 'I_X3_B', 'I_X3_C', 'X2_I_X3_B', 'X2_I_X3_C']]
y_train = train_df['Y']
X_train_model1 = sm.add_constant(X_train_model1)

# Model 2 (Reduced model)
X_train_model2 = train_df[['X1', 'X2', 'I_X3_B', 'I_X3_C']]
X_train_model2 = sm.add_constant(X_train_model2)

# Model 3 (Large model)
X_train_model3 = train_df[['X1', 'X2', 'X1_squared', 'X2_squared', 'I_X3_B', 'I_X3_C', 
                           'X2_I_X3_B', 'X2_I_X3_C', 'X1_I_X3_B', 'X1_I_X3_C']]
X_train_model3 = sm.add_constant(X_train_model3)

# Model 1
model1 = sm.OLS(y_train, X_train_model1).fit()

# Model 2
model2 = sm.OLS(y_train, X_train_model2).fit()

# Model 3
model3 = sm.OLS(y_train, X_train_model3).fit()


def plot_residuals(fitted_model, X, y, title):
    y_pred = fitted_model.predict(X)
    residuals = y - y_pred
    
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))
    
    # Residual plot
    ax[0].scatter(y_pred, residuals)
    ax[0].axhline(0, color='gray', linestyle='--')
    ax[0].set_xlabel('Predicted Y')
    ax[0].set_ylabel('Residuals')
    ax[0].set_title(f'Residual Plot: {title}')
    
    # Q-Q plot (for checking normality of residuals)
    sm.qqplot(residuals, line ='45', fit=True, ax=ax[1])
    ax[1].set_title(f'Q-Q Plot: {title}')
    
    plt.tight_layout(rect=[0, 0, 1, 0.95]) 
    fig.suptitle(title, fontsize=16)
    plt.show()

plot_residuals(model1, X_train_model1, y_train, 'Model 1 (Correct Predictors)')
plot_residuals(model2, X_train_model2, y_train, 'Model 2 (Reduced Model)')
plot_residuals(model3, X_train_model3, y_train, 'Model 3 (Large Model)')

X_test_model1 = test_df[['X1', 'X2', 'I_X3_B', 'I_X3_C', 'X2_I_X3_B', 'X2_I_X3_C']]
X_test_model1 = sm.add_constant(X_test_model1)
y_test = test_df['Y']

X_test_model2 = test_df[['X1', 'X2', 'I_X3_B', 'I_X3_C']]
X_test_model2 = sm.add_constant(X_test_model2)

X_test_model3 = test_df[['X1', 'X2', 'X1_squared', 'X2_squared', 'I_X3_B', 'I_X3_C', 
                         'X2_I_X3_B', 'X2_I_X3_C', 'X1_I_X3_B', 'X1_I_X3_C']]
X_test_model3 = sm.add_constant(X_test_model3)
```

```python
'''Copied from the homework directions'''
import statsmodels.api as sm
import pandas as pd

def stepwise_selection_AIC(X, Y, initial_list=[], threshold_in=0.01, threshold_out=0.01, verbose=True):
    
    included = list(initial_list)
    best_aic = np.inf

    while True:
        changed = False
        excluded = list(set(X.columns) - set(included))
        aic_with_candidates = pd.Series(index=excluded, dtype=float)

        for new_column in excluded:
            model = sm.OLS(Y, sm.add_constant(X[included + [new_column]])).fit()
            aic_with_candidates[new_column] = model.aic

        best_candidate_aic = aic_with_candidates.min()
        if best_candidate_aic < best_aic:
            best_aic = best_candidate_aic
            best_feature = aic_with_candidates.idxmin()
            included.append(best_feature)
            changed = True

            if verbose:
                print(f'Add {best_feature} with AIC {best_candidate_aic}')
        
        model = sm.OLS(Y, sm.add_constant(X[included])).fit()
        current_aic = model.aic
        aic_without_candidates = pd.Series(index=included, dtype=float)

        for candidate in included:
            model = sm.OLS(Y, sm.add_constant(X[included].drop(columns=[candidate]))).fit()
            aic_without_candidates[candidate] = model.aic

        worst_candidate_aic = aic_without_candidates.min()
        if worst_candidate_aic < best_aic:
            best_aic = worst_candidate_aic
            worst_feature = aic_without_candidates.idxmin()
            included.remove(worst_feature)
            changed = True
            
            if verbose:
                print(f'Remove {worst_feature} with AIC {worst_candidate_aic}')

        if not changed:
            break
    return included

'''So I made X test and train data based off od each of the models. Because I used Model 3 (large model) 
as instructedin the task above this, I will also use Model 3 train and testing data here.'''

selected_features = stepwise_selection_AIC(X_train_model3, y_train)

X_train_selected = sm.add_constant(X_train_model3[selected_features])
model_selected = sm.OLS(y_train, X_train_selected).fit()  #y_train is from my above code where I made the test data

print("\nSelected features:", selected_features, '\n')
print(model_selected.summary())

X_test_selected = sm.add_constant(X_test_model3[selected_features])

Y_pred_test = model_selected.predict(X_test_selected)
mse_test = mean_squared_error(y_test, Y_pred_test)               #y_test is from my above code where I made the test data
print(f"\nMean Squared Error (MSE) on the testing set: {mse_test}")

    
```

```python
'''Using the Model 3 again, because it was used in the previous tasks'''
# Assuming you've defined y_train and Xtrain_model3 (the one I created for the OG SFFS task) correctly

model_selected = sm.OLS(y_train, Xtrain_model3).fit()
coefficients = model_selected.params

n_realizations = 10
new_Y = pd.DataFrame()

original_residuals = model_selected.resid
std_error = original_residuals.std()

for i in range(n_realizations):
    random_errors = np.random.normal(loc=0, scale=std_error, size=len(X_train_model3))
    
    X_with_constant = sm.add_constant(X_train_model3)
    
    new_Y_realization = np.dot(X_with_constant, coefficients) + random_errors
    new_Y[f'Realization_{i + 1}'] = new_Y_realization

print(new_Y.head(15))
```
